```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
contra_costa_covid_2020=pd.read_csv('contra_costa_covid_2020.csv')  
alameda_covid_2020=pd.read_csv('alameda_covid_2020.csv')  


contra_costa_2020_spending=pd.read_csv('contra_costa_2020_spending.csv')  
contra_costa_2019_spending=pd.read_csv('contra_costa_2019_spending.csv')  
alameda_2020_spending=pd.read_csv('alameda_2020_spending.csv')  
alameda_2019_spending=pd.read_csv('alameda_2019_spending.csv') 
contra_costa_2020_spending['Spending']=contra_costa_2020_spending['Spending']/1000000
contra_costa_2019_spending['Spending']=contra_costa_2019_spending['Spending']/1000000
alameda_2020_spending['Spending']=alameda_2020_spending['Spending']/1000000
alameda_2019_spending['Spending']=alameda_2019_spending['Spending']/1000000
```

```python
plotdata = pd.DataFrame({
    "2019":contra_costa_2019_spending['Spending'].values,
    "2020":contra_costa_2020_spending['Spending'].values,
    }, 
    index=contra_costa_2019_spending['Week'].values
)
plotdata.plot(kind="bar",figsize=(25,15), fontsize=18)
plt.xlabel("Week Number",fontsize=20)
plt.ylabel("Weekly spending ($million)",fontsize=20)
plt.title("Contra Costa County weekly spending",fontsize=30)
plt.legend(["2019", "2020"], prop={"size":20})
plt.savefig("Contra_Costa_Spending.png", dpi = 300, transparent=True, bbox_inches="tight")
```

```python
plotdata = pd.DataFrame({
    "2019":alameda_2019_spending['Spending'].values,
    "2020":alameda_2020_spending['Spending'].values,
    }, 
    index=alameda_2019_spending['Week'].values
)
plotdata.plot(kind="bar",figsize=(25,15), fontsize=18)
plt.xlabel("Week Number",fontsize=20)
plt.ylabel("Weekly spending ($million)",fontsize=20)
plt.title("Alameda County weekly spending",fontsize=30)
plt.legend(["2019", "2020"], prop={"size":20})
plt.savefig("Alameda_Spending_full.png", dpi = 300, transparent=True, bbox_inches="tight")
```

```python
plotdata = pd.DataFrame({
    "2019":alameda_2019_spending['Spending'].values[34:],
    "2020":alameda_2020_spending['Spending'].values[34:],
    }, 
    index=alameda_2019_spending['Week'].values[34:]
)
plotdata.plot(kind="bar",figsize=(25,15), fontsize=18)
plt.xlabel("Week Number",fontsize=20)
plt.ylabel("Weekly spending ($million)",fontsize=20)
plt.title("Alameda County weekly spending",fontsize=30)
plt.legend(["2019", "2020"], prop={"size":20})
plt.savefig("Alameda_Spending_short.png", dpi = 300, transparent=True, bbox_inches="tight")
```

```python
plotdata = pd.DataFrame({
    "Contra Costa County":contra_costa_covid_2020['Covid'].values,
    "Alameda County":alameda_covid_2020['Covid'].values,
    }, 
    index=contra_costa_2019_spending['Week'].values
)
plotdata.plot(kind="bar",figsize=(25,15), fontsize=18)
plt.xlabel("Week Number",fontsize=20)
plt.ylabel("Weekly Covid Cases",fontsize=20)
plt.title("Contra Costa County weekly spending",fontsize=30)
plt.legend(["Contra Costa County", "Alameda County"], prop={"size":20})
plt.savefig("Covid Cases.png", dpi = 300, transparent=True, bbox_inches="tight")
```
