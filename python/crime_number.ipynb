```python
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
crime=pd.read_csv('crime_labels.csv')
```

```python
crimes=crime[['Email Text','Date','Zip Code']]
```

```python
crime_list=['Arson', 'Assault', 'Burglary', 'Disturbing the Peace', 'Drugs / Alcohol Violations', 'Fraud', 'Homicide', 'Motor Vehicle Theft', 'Robbery', 'Sex Crimes', 'Theft / Larceny', 'Vandalism', 'Vehicle Break-In / Theft', 'Weapons']
crimes[crime_list] = 0
crimes=crimes.dropna()
crimes['Zip Code']=crimes['Zip Code'].astype(int)
```

```python
for i in crime_list:
    if i=='Theft / Larceny' or i=='Drugs / Alcohol Violations':
        crimes[i]=crimes['Email Text'].str.count(r"\b{}\b".format(i))
    else:
        crimes[i]=crimes['Email Text'].str.count(r"\b{}\b".format(i))-1
```

```python
crimes['Total crimes']=np.sum(crimes[crime_list],axis=1)
```

```python
crimes_grouped=crimes.groupby('Zip Code').sum()
```
