```python
#2019 stops

import sys
import psycopg2
import pickle


#==============================================================================
# Main function
#==============================================================================
def main(argv):
    tbl = 'actransit'
    fname = 'active_stops_2019.pickle'
    conn = psycopg2.connect("dbname=transit user=postgres", password="postgres")
    cur = conn.cursor()
    
    active_stops_2020=[]
    active_stops_2019=[]
    
    stmt= "SELECT DISTINCT stop_id FROM actransit"
    stmt += " WHERE act_stop_time >= TO_DATE('2019-01-01', 'YYYY-MM-DD') AND act_stop_time <=  TO_DATE('2019-12-31', 'YYYY-MM-DD');"
    print(stmt)
    cur.execute(stmt)
    
    result = cur.fetchall()
    
    stops = []
    print(result)

    for record in result:
        stops.append(record[0])
         
    active_stops_2019.append(stops)

    conn.commit()
    cur.close()
    conn.close()


    with open(fname, 'wb') as f:
        pickle.dump([active_stops_2019], f)
        
    f.close






if __name__ == "__main__":
    main(sys.argv)
```

```python
#passengers on per stop pay day of week 2020
import sys
import psycopg2
import pickle
import csv


#==============================================================================
# Main function
#==============================================================================
def main(argv):
    tbl = 'actransit'
    fname = 'stops_by_passenger_2020.pickle'
    conn = psycopg2.connect("dbname=transit user=postgres", password="postgres")
    cur = conn.cursor()
    week = ['0', '1', '2', '3', '4', '5', '6']

    dow_passenger=[]
    dow_2020=[]
    stops_2020=[]
    for dow in week:
        stmt = "SELECT DATE(act_stop_time), stop_id, sum(psgr_on) as passengers from "
        stmt += tbl
        stmt += " WHERE EXTRACT(DOW FROM act_stop_time) = " + dow
        stmt += " AND act_stop_time >= TO_DATE('2020-01-01', 'YYYY-MM-DD') AND act_stop_time <=  TO_DATE('2020-12-31', 'YYYY-MM-DD') "
        stmt += " GROUP BY DATE(act_stop_time), stop_id "
        stmt += " ORDER BY DATE(act_stop_time), passengers;"
    
        cur.execute(stmt)
    
        result = cur.fetchall()
        date=[]
        stop_id=[]
        passenger=[]
        
        for record in result:
            date.append(record[0])
            stop_id.append(record[1])
            passenger.append(record[2])
        dow_2020.append(date)
        stops_2020.append(stop_id)
        dow_passenger.append(passenger)

    conn.commit()
    cur.close()
    conn.close()


    with open(fname, 'wb') as f:
        pickle.dump([dow_2020, stops_2020, dow_passenger], f)
        
    f.close






if __name__ == "__main__":
    main(sys.argv)
```
